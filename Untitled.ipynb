```python
import subprocess
import os

# === CONFIGURATION ===
video_url = "https://www.youtube.com/watch?v=Nl3is7n3PKw&t=4s"  # Replace with your target video URL
output_video = "/teamspace/studios/this_studio/Validatition_video.mp4"  # The final mp4 output
output_audio = "/teamspace/studios/this_studio/Validatition/audio.wav"  # The extracted audio file
cookies_file = "/teamspace/studios/this_studio/youtube.com_cookies.txt"  # Path to your exported browser cookies

# Ensure the output directory exists
output_audio_dir = os.path.dirname(output_audio)
os.makedirs(output_audio_dir, exist_ok=True)

# === STEP 1: Download video with yt-dlp using cookies ===
print("Downloading video...")
download_cmd = [
    "yt-dlp",
    "--cookies", cookies_file,
    "-f", "bestvideo+bestaudio/best",
    "--merge-output-format", "mp4",  # Force mp4 output
    "-o", output_video,
    video_url
]

try:
    subprocess.run(download_cmd, check=True)
    print("Video downloaded successfully.")
except subprocess.CalledProcessError as e:
    print("Failed to download video:", e)
    exit(1)

# === STEP 2: Extract WAV audio using ffmpeg ===
print("Extracting audio to WAV...")
extract_audio_cmd = [
    "ffmpeg",
    "-i", output_video,  # Using the correct video file path
    "-vn",               # No video
    "-acodec", "pcm_s16le",  # WAV format
    "-ar", "44100",      # Sampling rate
    "-ac", "2",          # Stereo
    "-map", "0:a:0",     # Ensure we are using the correct audio stream
    output_audio
]

try:
    subprocess.run(extract_audio_cmd, check=True)
    print("Audio extracted successfully:", output_audio)
except subprocess.CalledProcessError as e:
    print("Failed to extract audio:", e)
    exit(1)
```
